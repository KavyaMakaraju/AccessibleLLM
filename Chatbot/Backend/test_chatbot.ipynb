```python

from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
from transformers import pipeline
from langchain.prompts import PromptTemplate
import torch
import nest_asyncio
from pyngrok import ngrok
import uvicorn

# Define the template for the chatbot
template = """ Answer the query from the human. Provide a single answer only. If you do not know the answer to the question, say "I don't know"

Context: You are a chatbot that is friendly and helpful. Your purpose is to behave like a friendly assistant that responds to the human query in a friendly manner.

Example Conversation:
Human: Hi!
Chatbot: Hello. How are you?
Real conversation starts from here. You should extract the Jack part ONLY.

Current conversation:
{history}

User: {input}

Jack: """
PROMPT = PromptTemplate(input_variables=["input", "history"], template=template)

# Sample function to simulate chat interactions
history = ""  # Make history a global variable

def generate_response(prompt: str, max_new_tokens: int = 128) -> str:
    # Assuming you have a preloaded Hugging Face pipeline for text generation
    generator = pipeline('text-generation', model='gpt2')  # Replace 'gpt2' with your model
    result = generator(prompt, max_new_tokens=max_new_tokens)
    return result[0]['generated_text']

def chat_interaction(user_input):
    global history  # Declare history as global inside the function

    # Format the prompt using the template
    formatted_prompt = PROMPT.format(input=user_input, history=history)

    # Generate the response using the predefined generate_response function
    chatbot_response = generate_response(formatted_prompt)

    # Add the user input and chatbot response to the history
    history += f"User: {user_input}\nJack: {chatbot_response}\n"

    return chatbot_response

# FastAPI app initialization
app = FastAPI()

# CORS middleware
origins = ["*"]
app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

class ChatInput(BaseModel):
    user_input: str

@app.post("/chat")
async def chat(conversation: ChatInput):
    # Generate the response using the chat_interaction function
    chatbot_response = chat_interaction(conversation.user_input)

    return {"response": chatbot_response, "history": history}

@app.get("/")
async def root():
    return {"hello": "world"}

if __name__ == "__main__":
    # Set up ngrok
    ngrok.set_auth_token("2dQXsslqnJ7w2f9QPyQzfR8Uwff_7KDN7JWR66Vbvq1ieCEin")
    ngrok_tunnel = ngrok.connect(8000)
    print('Public URL:', ngrok_tunnel.public_url)

    # Apply nest_asyncio
    nest_asyncio.apply()

    # Run the FastAPI app with uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
