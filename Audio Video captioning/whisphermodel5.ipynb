```python
import whisper
import os
import csv
from difflib import SequenceMatcher

# Load the CSV file
csv_file = "/content/drive/MyDrive/speech-to-text-dataset/cv-other-train.csv"  # Change this to your CSV file path
transcriptions = {}
with open(csv_file, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        filename, transcription = row
        transcriptions[filename] = transcription.lower()


model = whisper.load_model("medium")


audio_folder = "/content/drive/MyDrive/speech-to-text-dataset/cv-other-train/"  # Change this to your audio files folder

# Initialize lists to store evaluation metrics
accuracies = []
precisions = []
recalls = []


for filename in os.listdir(audio_folder):
    if filename.endswith(".mp3"):  # Assuming all files are MP3 format
        file_path = os.path.join(audio_folder, filename)

        # Transcribe audio
        try:
            result = model.transcribe(file_path)
            transcription = result["text"].lower()
        except Exception as e:
            print(f"Error transcribing {filename}: {e}")
            continue


        ground_truth = transcriptions.get(filename, "")

        matcher = SequenceMatcher(None, transcription.split(), ground_truth.split())
        matching_words = sum(match.size for match in matcher.get_matching_blocks())


        precision = (matching_words / len(transcription.split())) * 100 if len(transcription.split()) > 0 else 0
        recall = (matching_words / len(ground_truth.split())) * 100 if len(ground_truth.split()) > 0 else 0

        total_words = len(ground_truth.split()) if ground_truth else 1  # Avoid division by zero
        accuracy = (matching_words / total_words) * 100


        accuracies.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)


        print(f"File: {filename}")
        print(f"Transcription: {transcription}")
        print(f"Ground Truth: {ground_truth}")
        print(f"Accuracy: {accuracy:.2f}%")
        print(f"Precision: {precision:.2f}%")
        print(f"Recall: {recall:.2f}%")
        print("--------------------")

print("Average Metrics:")
print(f"Average Accuracy: {sum(accuracies)/len(accuracies):.2f}%")
print(f"Average Precision: {sum(precisions)/len(precisions):.2f}%")
print(f"Average Recall: {sum(recalls)/len(recalls):.2f}%")
```

```python
import whisper
import os
import csv
import pandas as pd
from difflib import SequenceMatcher
from jiwer import wer
from jiwer import cer
import re

csv_file = "/content/drive/MyDrive/speech-to-text-dataset/cv-other-train.csv"  # Change this to your CSV file path
transcriptions = {}
with open(csv_file, 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        filename, transcription = row
        transcriptions[filename] = transcription.lower()

model = whisper.load_model("medium")

audio_folder = "/content/drive/MyDrive/speech-to-text-dataset/cv-other-train/"  # Change this to your audio files folder

evaluation_data = []

for filename in os.listdir(audio_folder):
    if filename.endswith(".mp3"):
        file_path = os.path.join(audio_folder, filename)

        try:
            result = model.transcribe(file_path)
            transcription = result["text"].lower()
        except Exception as e:
            print(f"Error transcribing {filename}: {e}")
            continue

        ground_truth = transcriptions.get(filename, "")


        transcription_cleaned = re.sub(r'[^\w\s]', '', transcription)
        ground_truth_cleaned = re.sub(r'[^\w\s]', '', ground_truth)
        transcription_cleaned = ' '.join(transcription_cleaned.split())
        ground_truth_cleaned = ' '.join(ground_truth_cleaned.split())

        matcher = SequenceMatcher(None, transcription_cleaned, ground_truth_cleaned)
        matching_words = sum(match.size for match in matcher.get_matching_blocks())

        precision = (matching_words / len(transcription_cleaned)) * 100 if len(transcription_cleaned) > 0 else 0
        recall = (matching_words / len(ground_truth_cleaned)) * 100 if len(ground_truth_cleaned) > 0 else 0

        total_words = len(ground_truth_cleaned) if ground_truth_cleaned else 1  # Avoid division by zero
        accuracy = (matching_words / total_words) * 100

        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        wer_score = wer(ground_truth_cleaned, transcription_cleaned)
        cer_score = cer(ground_truth_cleaned, transcription_cleaned)

        evaluation_data.append({
            'Filename': filename,
            'Ground Truth': ground_truth,
            'Transcription': transcription,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-score': f1_score,
            'WER': wer_score,
            'CER': cer_score
        })

evaluation_df = pd.DataFrame(evaluation_data)

output_excel_file = "evaluation_results.xlsx"
evaluation_df.to_excel(output_excel_file, index=False)
```

```python

overall_accuracy = evaluation_df['Accuracy'].mean()
overall_precision = evaluation_df['Precision'].mean()
overall_recall = evaluation_df['Recall'].mean()
overall_f1_score = evaluation_df['F1-score'].mean()
overall_wer = evaluation_df['WER'].mean()
overall_cer = evaluation_df['CER'].mean()

#overall metrics
print("Overall Metrics:")
print(f"Overall Accuracy: {overall_accuracy:.2f}%")
print(f"Overall Precision: {overall_precision:.2f}%")
print(f"Overall Recall: {overall_recall:.2f}%")
print(f"Overall F1-score: {overall_f1_score:.2f}")
print(f"Overall WER: {overall_wer:.2f}")
print(f"Overall CER: {overall_cer:.2f}")
```
